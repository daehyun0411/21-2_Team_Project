## 날짜별/산지별 그룹화

```python
cab = cab.groupby(['SAN_NM','SALEDATE'])['TOT_QTY','TOT_AMT'].sum().reset_index()
```

```python
cab
```

## 년도별 주산지?

```python
joosan_dict = dict()

for i in (2015,2016,2017,2018,2019,2020,2021):
    cab_year = cab[cab['SALEDATE'].dt.year == i]
    joosan_list = cab_year.groupby(['SAN_NM'])['TOT_QTY'].sum().sort_values(ascending=False).reset_index()
    joosan_dict[i] = joosan_list.iloc[0]
```
