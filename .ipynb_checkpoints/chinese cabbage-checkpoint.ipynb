```python
import pandas as pd
import numpy as np
```

# 배추 - 도매시장

```python
cab = pd.read_excel('배추.xlsx', parse_dates=['일자'])
```

```python
cab
```

```python
cab.info()
```

## 법인

```python
cab.drop(columns=['법인'], axis=1, inplace=True)
```

## 출하지

```python
cab['출하지'].unique()
```

```python
drop_index = cab[(cab['출하지'] == '-')|(cab['출하지'] == '수입산')].index
cab.drop(index = drop_index, axis=0, inplace=True)
```

```python
name_split = cab['출하지'].str.split(" ")
```

```python
cab['출하지'] = name_split.str.get(0)
```

```python
cab['출하지'].replace('광주','광주광역시',inplace=True)
cab['출하지'].replace('대구','대구광역시',inplace=True)
cab['출하지'].replace('대전','대전광역시',inplace=True)
cab['출하지'].replace('부산','부산광역시',inplace=True)
cab['출하지'].replace('서울','서울특별시',inplace=True)
cab['출하지'].replace('울산','울산광역시',inplace=True)
cab['출하지'].replace('인천','인천광역시',inplace=True)
```

```python
cab['출하지'].unique()
```

## 금액/물량

```python
cab[(cab['금액(원)'] <= 0)|(cab['물량(kg)'] <=0)]
```

```python
drop_index2 = cab[(cab['금액(원)'] <= 0)|(cab['물량(kg)'] <=0)].index
cab.drop(index= drop_index2, axis=0, inplace=True)
```

## 단위당금액

```python
cab['단위당금액'] = round(cab['금액(원)']/cab['물량(kg)'],0)
```

```python
cab['단위당금액'] = cab['단위당금액'].astype(int)
```

```python
cab
```

## 저장

```python
cab.to_excel('배추_수정.xlsx', index=False)
```
