```python
import pandas as pd
import numpy as np
```

# 갓

```python
mus = pd.read_excel('갓.xlsx', parse_dates=['일자'])
```

```python
mus
```

```python
mus.info()
```

## 법인

```python
mus.drop(columns=['법인'], axis=1, inplace=True)
```

## 출하지

```python
mus['출하지'].unique()
```

```python
drop_index = mus[(mus['출하지'] == '-')].index
mus.drop(index = drop_index, axis=0, inplace=True)
```

```python
name_split = mus['출하지'].str.split(" ")
mus['출하지'] = name_split.str.get(0)
```

```python
mus['출하지'].replace('광주','광주광역시',inplace=True)
mus['출하지'].replace('대구','대구광역시',inplace=True)
mus['출하지'].replace('대전','대전광역시',inplace=True)
mus['출하지'].replace('부산','부산광역시',inplace=True)
mus['출하지'].replace('서울','서울특별시',inplace=True)
mus['출하지'].replace('울산','울산광역시',inplace=True)
mus['출하지'].replace('인천','인천광역시',inplace=True)
```

```python
mus['출하지'].unique()
```

## 물량/금액

```python
mus[(mus['금액(원)'] <= 0)|(mus['물량(kg)'] <=0)]
```

```python
drop_index2 = mus[(mus['금액(원)'] <= 0)|(mus['물량(kg)'] <=0)].index
mus.drop(index= drop_index2, axis=0, inplace=True)
```

## 단위당금액

```python
mus['단위당금액'] = round(mus['금액(원)']/mus['물량(kg)'],0)
```

```python
mus['단위당금액'] = mus['단위당금액'].astype(int)
```

## 저장

```python
mus.to_excel('갓_수정.xlsx', index=False)
```
